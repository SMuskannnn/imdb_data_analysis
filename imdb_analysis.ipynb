# Data preprocessing using pandas

### Importing required modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Inspecting the data

```python
# load dataset as dataframe
df = pd.read_csv('imdb_top_1000.csv')
```

```python
# Set options to display all columns in a single line
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
#display first few rows
df.head(30)
```

```python
# checking all the unique values
def get_unique(col):
    return col.unique()

for col in df.columns:
    unique_values = get_unique(df[col])
    print(f"Unique values for column '{col}': {unique_values}")
    print()
```

```python
#check for missing values
print(df.isnull().sum())
```

```python
# making sure that all the column names are in lowercase
df.columns = [x.lower() for x in df.columns]
df.columns
```

```python
#checking for datatypes
df.dtypes
```

### adjusting the data

```python
# gross column
df['gross'].str.strip() 
df['gross'] = df['gross'].str.replace(',','')  
df['gross'] = pd.to_numeric(df['gross'], errors = 'coerce').astype(float)
df['gross'] = df['gross']/1000000
df.rename(columns = {'gross':'gross(million)'},inplace = True)
df['gross(million)']
```

```python
#released_year
df['released_year'] = df['released_year'].replace('PG','N/A')
```

```python
# runtime
df['runtime'] = df['runtime'].str.extract('(\d+)').astype(float)
df.rename(columns = {'runtime':'runtime(minutes)'},inplace = True)
df['runtime(minutes)']
```

```python
df.columns
```

### Handling missing values

```python
# visualizing the null values throughout the dataset
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap='coolwarm', cbar=True)
plt.title('Null Values Heatmap')
plt.show()
```

```python
df['gross(million)']= df['gross(million)'].replace(np.nan,'N/A')
df = df.fillna({'meta_score':'N/A','certificate':'N/A'})
```

### Duplicate values

```python
if (df.duplicated().any()):
    print("Duplicates")
else:
    print("NO Duplicates")
```

# Data Analysis

### average gross earning of top 100 high gross earning directors

```python
# Convert 'Gross Earnings' to numeric (remove any non-numeric characters if needed)
df['gross(million)'] = pd.to_numeric(df['gross(million)'].replace('[\$,]', '', regex=True), errors='coerce')

# Calculate average gross earnings per movie for each director
avg_earnings_per_director = df.groupby('director')['gross(million)'].mean().reset_index()
avg_earnings_per_director
```

```python
# Sort the DataFrame by average gross earnings for better visualization
avg_earnings_per_director = avg_earnings_per_director.sort_values(by='gross(million)', ascending=False)

#for top 100 high gross earning directors
high_earnings_director = avg_earnings_per_director.head(100)

# Bar chart
plt.figure(figsize=(12, 6))
plt.bar(high_earnings_director['director'], high_earnings_director['gross(million)'], color='skyblue')
plt.xlabel('Director')
plt.ylabel('Average Gross Earnings per Movie (Million)')
plt.title('Average Gross Earnings per Movie for Each Director')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

### Impact of director on gross earning

#### distribution of earnings throughout the movies using box plot

```python
#top 100 directors according to their average gross earning
df['mean_earnings'] = df['director'].map(avg_earnings_per_director.set_index('director')['gross(million)'])
df_100 = df.sort_values(by='mean_earnings', ascending=False).head(100)
# Box plot to visualize distribution of earnings for each director
plt.figure(figsize=(12, 6))
sns.boxplot(x='director', y='gross(million)', data=df_100)
plt.xticks(rotation=90)
plt.show()
```

### genre popularity over years

```python
# Split the genres and stack them to create a new DataFrame
#since the genre are comma seperated in one column
genre_df = df['genre'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('genre')
genre_df = genre_df.str.lstrip()
genre_df
```

```python
# Merge the new DataFrame with the original DataFrame df
df_split = df.drop('genre', axis=1).merge(genre_df, left_index=True, right_index=True)
df_split[['released_year','genre']]
```

```python
# Calculate the frequency of each genre per year
genre_frequency = df_split.groupby(['released_year', 'genre']).size().reset_index(name='Frequency')

# Identify genres that have gained or lost popularity over time
pivot_table = genre_frequency.pivot_table(index='released_year', columns='genre', fill_value=0)
genre_changes = pivot_table.diff().fillna(0)
genre_changes = genre_changes.sort_values(by = 'released_year',ascending = False)
# Print the results
print("Genre Frequency per Year:")
print(genre_frequency)

print("\nGenre Changes Over Time:")
print(genre_changes.head(10))

```

```python
# Plotting the heatmap for the first 10 recent years
plt.figure(figsize=(10, 6))
sns.heatmap(genre_changes.head(10), cmap='coolwarm', annot=True, fmt='.0f', linewidths=.5)
plt.title('Genre Changes Over Time (First 10 recent years)')
plt.xlabel('Genre')
plt.ylabel('Year')
plt.show()
```

### Correlation between IMDB Ratings and Gross Earnings:

```python
# Convert 'imdb_rating' to numeric (remove any non-numeric characters if needed)
df['imdb_rating'] = pd.to_numeric(df['imdb_rating'].replace('[\$,]', '', regex=True), errors='coerce')

# Create a scatter plot to show the correlation between IMDB ratings and gross earnings
plt.figure(figsize=(8, 4))
sns.scatterplot(x='imdb_rating', y='gross(million)', data=df)
plt.xlabel('IMDB Rating')
plt.ylabel('Gross Earnings (Million)')
plt.title('Correlation between IMDB Ratings and Gross Earnings')
plt.show()

# Calculate correlation coefficient
correlation_coefficient = df['imdb_rating'].corr(df['gross(million)'])
print(f'Correlation Coefficient: {correlation_coefficient}')
```

### Impact of Movie Length on Ratings or Earnings:

```python
# Convert 'duration' to numeric (remove any non-numeric characters if needed)
df['runtime(minutes)'] = pd.to_numeric(df['runtime(minutes)'].replace('[\$,]', '', regex=True), errors='coerce')

# Set up a 1x2 grid for subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Scatter plot for the impact of movie length on IMDB ratings
sns.scatterplot(x='runtime(minutes)', y='imdb_rating', data=df, ax=axes[0])
axes[0].set_xlabel('Movie Length (minutes)')
axes[0].set_ylabel('IMDB Rating')
axes[0].set_title('Impact of Movie Length on IMDB Ratings')

# Scatter plot for the impact of movie length on gross earnings
sns.scatterplot(x='runtime(minutes)', y='gross(million)', data=df, ax=axes[1])
axes[1].set_xlabel('Movie Length (minutes)')
axes[1].set_ylabel('Gross Earnings (Million)')
axes[1].set_title('Impact of Movie Length on Gross Earnings')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()

# Calculate correlation coefficients
correlation_coefficient_ratings_length = df['runtime(minutes)'].corr(df['imdb_rating'])
correlation_coefficient_earnings_length = df['runtime(minutes)'].corr(df['gross(million)'])

print(f'Correlation Coefficient (Ratings vs. Length): {correlation_coefficient_ratings_length}')
print(f'Correlation Coefficient (Earnings vs. Length): {correlation_coefficient_earnings_length}')
```

### classification of movies based on ratings:

```python
#class= ['Poor', 'Below Average', 'Average', 'Excellent']
def classify(rating):
    if rating >= 8.0:
        return "Excellent"
    elif rating >= 6.0:
        return 'Average'
    elif rating >= 4.0:
        return 'Below Average'
    else:
        return 'Poor'

# Apply the classify function to create a new column 'rating_class'
df['rating_class'] = df['imdb_rating'].apply(classify)

# Display the DataFrame with the new 'rating_class' column
df[['series_title', 'imdb_rating', 'rating_class']].head()
```

```python
# Create a countplot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.countplot(x='rating_class', data=df, order=['Poor', 'Below Average', 'Average', 'Excellent'], palette='viridis')
plt.title('Distribution of Movies Based on Rating Classes')
plt.xlabel('Rating Class')
plt.ylabel('Number of Movies')
plt.show()
```
